```python
capture_mc = cv2.VideoCapture(path_mcgill)
```

```python
# Assuming capture_mc is already defined and is an OpenCV video capture object
frame_mc = 0  # Initialize the frame counter

while True:
    true, frame = capture_mc.read()

    if true:
        # Correct string formatting for the filename
        filename = '/content/drive/MyDrive/Colab Notebooks/mcgill_frames/frame{}.jpg'.format(frame_mc)
        cv2.imwrite(filename, frame)
        frame_mc += 1  # Increment the frame counter
    else:
        break

capture_mc.release()  # Release the video capture object
```

### Number of Padestrien passed.

```python
import os

# Path to the labels directory produced by YOLOv5
labels_dir = '/content/yolov5/yolov5/runs/detect/exp/labels'

# Class index for 'person' in COCO dataset is typically 0
person_class_index = 0

# Initialize a counter for pedestrians
total_pedestrians = 0

# List all the label files in the directory
label_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]

# Loop through all label files and count the pedestrians
for label_file in label_files:
    file_path = os.path.join(labels_dir, label_file)
    with open(file_path, 'r') as file:
        for line in file.readlines():
            class_index = int(line.split()[0])
            if class_index == person_class_index:
                total_pedestrians += 1

# Print the total number of pedestrians detected
print(f"Total number of pedestrians detected: {total_pedestrians}")
```
